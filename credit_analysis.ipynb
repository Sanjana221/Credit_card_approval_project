```python
from sklearn import preprocessing 
label_encoder = preprocessing.LabelEncoder() 
data['Car_Owner']= label_encoder.fit_transform(data['Car_Owner'])
data['Propert_Owner']= label_encoder.fit_transform(data['Propert_Owner'])
data['EDUCATION']= label_encoder.fit_transform(data['EDUCATION'])
data['GENDER']= label_encoder.fit_transform(data['GENDER'])
data['Car_Owner']= label_encoder.fit_transform(data['Car_Owner'])
data['Type_Income']= label_encoder.fit_transform(data['Type_Income'])
data['Marital_status']= label_encoder.fit_transform(data['Marital_status'])
data['Housing_type']= label_encoder.fit_transform(data['Housing_type'])
```

```python
new_data = data[['Debt', 'BankCustomer','PriorDefault','Employed', 'CreditScore','YearsEmployed',
         'Income', 'Approved']]
```

```python
data = data[['Car_Owner', 'Propert_Owner','Annual_income','EDUCATION', 'label']]
```

```python
data.head()
```

```python
x = data.iloc[:, : -1]
y = data.iloc[:, -1]
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.head()
```

```python
from sklearn.preprocessing  import StandardScaler
scaler =   StandardScaler()
scaled_x_train = scaler.fit_transform(x_train)
scaled_x_test = scaler.fit_transform(x_test)
```

```python
from sklearn.metrics import accuracy_score
def model_performance(prediction):
    print(f"Accuracy score of model is {accuracy_score(y_test,prediction)}")
```

```python
from sklearn.linear_model import LogisticRegression
myModel = LogisticRegression(solver='lbfgs', max_iter=1000)
myModel.fit(scaled_x_train,y_train)
```

```python
Predictions = myModel.predict(scaled_x_test)
```

```python
model_performance(Predictions)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn import svm
svm = svm.SVC()
param_grid_svr = {"C" : [0.01,0.1,0.1,0.5,1] , 
                  "kernel" : ["linear","rbf","poly"]}
gridsvr = GridSearchCV(svm,param_grid_svr)
```

```python
gridsvr.fit(x_train,y_train)
```

```python
pred_svc = gridsvr.predict(x_test)
```

```python
model_performance(pred_svc)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
param_grid = {
    "criterion" :  ["gini","entropy"] ,
    "splitter" : ["best","random"] ,
    "max_depth" : [None,1,2,3,4,5],
    "min_samples_split" : [2,5,10] ,
    "min_samples_leaf" :  [1,2,4] 
}
grid_search_tree = GridSearchCV(estimator = DecisionTreeClassifier(), param_grid = param_grid)
```

```python
grid_search_tree.fit(scaled_x_train,y_train)
```

```python
pred_dec = grid_search_tree.predict(scaled_x_test)
```

```python
model_performance(pred_dec)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
n_estimators = [32,64,128,256,300,500,600] 
max_features = [2,3,4,5]
bootstrap = [True]
oob_score = [True,False]
```

```python
param_grid_rfc = {
   "n_estimators": n_estimators,
    "max_features" : max_features,
    "bootstrap" : bootstrap,
    "oob_score" : oob_score  
}
```

```python
grid_rfc = GridSearchCV(rfc , param_grid_rfc)
```

```python
grid_rfc.fit(scaled_x_train,y_train)
```

```python
grid_rfc.best_params_
```

```python
pred_rfc = grid_rfc.predict(scaled_x_test)
```

```python
model_performance(pred_rfc)
```

```python
from sklearn.ensemble import AdaBoostClassifier
ada_classifier = AdaBoostClassifier()
ada_param_grid = {
    "n_estimators" :[10,40,100],
    "learning_rate" : [0.01,0.1,1.10],
    "algorithm" : ["SAMME"]
}
ada_grid_search = GridSearchCV(ada_classifier,ada_param_grid,cv=2)
```
